```python
import pandas as pd
```

```python
df = pd.read_csv("vehicle_data.csv")
```

```python
df.head()
```

```python
df = df[df["vehicle_id"] == 0]
```

```python
df.head()
```

```python
# plot throttle, engine rpm, speed in one graph, with time in x axis
import matplotlib.pyplot as plt
```

```python
plt.xlabel("Time (s)")
plt.ylabel("Throttle (%)")
plt.plot(df["timestamp"], df["throttle"])
```

```python
plt.xlabel("Time (s)")
plt.ylabel("Speed (m/s)")
plt.plot(df["timestamp"], df["speed"])
```

```python
plt.xlabel("Time (s)")
plt.ylabel("Engine RPM")
plt.plot(df["timestamp"], df["engine_rpm"])
```

```python
plt.xlabel("Time (s)")
plt.ylabel("Acceleration (m/s^2)")
plt.plot(df["timestamp"], df["acceleration"])
```
